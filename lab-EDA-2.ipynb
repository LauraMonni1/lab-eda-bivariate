### Part 1: Analyzing Best-Seller Trends Across Product Categories

**Objective**: Understand the relationship between product categories and their best-seller status.

Insights after analysis:

- There are categories where BestSeller has a very low prevalence.

- However, the Bestseller distribution is statistically independent from the product categories.

**1. Crosstab Analysis**:

- Create a crosstab between the product category and the isBestSeller status.

- Are there categories where being a best-seller is more prevalent?

```python
crosstab = pd.crosstab(df["category"], df["isBestSeller"]).sort_values(by= True, ascending = False)
crosstab
```

```python
prop_crosstab = pd.crosstab(df["category"], df["isBestSeller"], normalize = "index").sort_values(by = True, ascending = False)
prop_crosstab
```

**2. Statistical Tests**:

- Conduct a Chi-square test to determine if the best-seller distribution is independent of the product category.
- Compute CramÃ©r's V to understand the strength of association between best-seller status and category.

```python
from scipy.stats import chi2_contingency
from scipy.stats.contingency import association
```

```python
chi2_statistic, p_value, _, _ = chi2_contingency(crosstab)
chi2_statistic, p_value
```

```python
association(crosstab, method = "cramer")
```

- Altaugh the chi2 test shows a p_value of 0, Cramer's V is only 0.122, indicating a quite weak association

**3. Visualizations**:

Visualize the relationship between product categories and the best-seller status using a stacked bar chart.

```python
crosstab.plot(kind="bar", stacked = True)
```

```python
def categorize_products(category):
    
    electronics = ["Speakers", "Players", "Technology", "Machines", "Hardware", "PC", "Computer", 
                   "Radio", "Radios" "CPUs", "3D", "Electrical", "Keyboards", "Laptop", "Smartwatches", 
                   "Electronics", "TV", "Scanners", "Headphones", "Hi-Fi", "Hard drive", "Phones", 
                   "Cameras", "Media", "GPS", "USB", "Microphones", "Alexa", "Telephones", 
                   "Printers", "Tablets", "Remote"]
    
    music = ["Music","Piano", "Drums", "Synthesisers", "Sound", "Bass", "Guitar", "DJ", "Karaoke"]
    
    home = ["Light", "Heating", "Coffee", "Bathroom", "Living", "Kitchen", "Stoves", "Fire", "Handmade"
           "Fragrances", "Furniture ", "Organisers", "Bedding", "Curtain", "Cookware", "Window", "Vacuums",
           "Lights", "Decorative", "Beds", "Household", "Frames", "Doormats", "Mirrors", "Clocks", 
            "Forniture", "indoor", "Cutting", "Tableware", "Cushions", "Painting", "Building", "Office", "Ironing"]
    
    sport = ["Motorbike", "Sports", "Ski", "Snowboards", "Skiing", "Hiking", "Outdoor", "Tennis", "Boating",
            "Cycling", "Basketball", "Darts", "Bowling", "Trampolines", "Billiard", "Cricket", "Golf",
            "Boxing", "Downhill", "Climbing", "Equestrian"]
    
    garden = ["Wind", "Plants", "Garden", "Bird", "Hobbies", "Power", "Agricultural", "Filters", "Gardening",
             "Pools", "Torches"]
    
    clothing = ["Boys", "Girls", "Men", "Baby", "Women"]
    
    beauty = ["Skin", "Body", "Beauty", "Make-up"]
    
    other = "Other"
    
    if category in electronics:
        return "Electronic and Media"
    elif category in music:
        return "Music"
    elif category in home:
        return "Home"
    elif category in sport:
        return "Sport"
    elif category in garden:
        return "Garden and Outdoor"
    elif category in clothing:
        return "Clothing"
    elif category in beauty:
        return "Beauty"
    else:
        return other
```

### Part 2: Exploring Product Prices and Ratings Across Categories and Brands

**Objective**: Investigate how different product categories influence product prices.

**0. Preliminary Step**: Remove outliers in product prices.

```python
def tukeys_test_outliers(data):
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1
    

    lower_q = Q1 - 1.5 * IQR
    upper_q = Q3 + 1.5 * IQR
    

    outliers = data[(data < lower_q) | (data > upper_q)] 
    
    return outliers
```

```python
data = df["price"]

outliers = tukeys_test_outliers(data)

df_no_outliers = df.iloc[~df.index.isin(outliers.index)]
```

```python
df_no_outliers.columns
```

```python
sns.boxplot(data=df_no_outliers['price'])
```

### Part 3: Investigating the Interplay Between Product Prices and Ratings

**1.Correlation Coefficients**:

Calculate the correlation coefficient between price and stars.
Is there a significant correlation between product price and its rating?

- The Spearman correlation coefficient between "category" and their rating ("stars") is - 0.32, indicating a weak negative relation between the two variables.

```python
corr_coeff = df["category"].corr(df["stars"], method = "spearman")  
# Spearmann correlation coefficient because data in column "stars" are ordinal, non-linear data 
#and this method is more efficient for this kind of data
corr_coeff
```
